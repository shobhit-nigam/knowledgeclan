```python
# 0 to 6 , early morning --> 4
# 6 to 12, morning --> 1
# 12 to 18, --> 2
# 18 to 24 --> 3

def assign_hour_label(h):
    if h >=0 and h < 6:
        return 4
    elif h >= 6 and h < 12:
        return 1
    elif h >= 12 and h < 18:
        return 2
    elif h >= 18 and h <=24:
        return 3
    
dfa["time_label"] = dfa["hr"].apply(assign_hour_label)
```

```python
train = dfa.sample(frac=0.8) 

test = dfa.loc[~dfa.index.isin(train.index)]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
list_col = list(dfa.columns)
```

```python
list_col.remove("cnt")
list_col.remove("casual")
list_col.remove("registered")
list_col.remove("dteday")
```

```python
model_reg = LinearRegression()
model_reg.fit(train[list_col], train["cnt"])
```

```python
# MSE 
```

```python
predictions_reg = model_reg.predict(test[list_col])

mse_reg = np.mean((predictions_reg - test["cnt"]) ** 2)
```

```python
print(f"mse with liner reg = {mse_reg}")
```

```python
from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor()
model_dt.fit(train[list_col], train["cnt"])

predictions_dt = model_dt.predict(test[list_col])

mse_dt = np.mean((predictions_dt - test["cnt"]) ** 2)

print(f"mse with Decision tree reg = {mse_dt}")
```

```python
from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor(min_samples_leaf=5)
model_dt.fit(train[list_col], train["cnt"])

predictions_dt = model_dt.predict(test[list_col])

mse_dt_leaf5 = np.mean((predictions_dt - test["cnt"]) ** 2)

print(f"mse with Decision tree reg (min samples leaf is 5) = {mse_dt_leaf5}")
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
model_rf.fit(train[list_col], train["cnt"])

predictions_rf = model_rf.predict(test[list_col])

mse_rf = np.mean((predictions_rf - test["cnt"]) ** 2)

print(f"mse with Decision tree reg (min samples leaf is 5) = {mse_rf}")
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(min_samples_leaf=5)
model_rf.fit(train[list_col], train["cnt"])

predictions_rf = model_rf.predict(test[list_col])

mse_rf_5 = np.mean((predictions_rf - test["cnt"]) ** 2)

print(f"mse with Decision tree reg (min samples leaf is 5) = {mse_rf_5}")
```
