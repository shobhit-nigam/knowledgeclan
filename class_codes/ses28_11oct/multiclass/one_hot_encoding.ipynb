```python
test = dfa.iloc[275:]
```

```python
# NA is positive (1) and europe & Asia cars considered negative (0)
# Europe is positive (1) and NA & Asia cars considered negative (0)
# Asia is positive (1) and europe & NA cars considered negative (0)
```

```python
# NA = 1
# Europe = 2
# Asia = 3
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
unique_origins = dfa["origin"].unique()
unique_origins.sort()
```

```python
models = {}
features = [x for x in train.columns if x.startswith("cyl") or x.startswith("year")]

for origin in unique_origins:
    model = LogisticRegression()
    
    X_train = train[features]
    y_train = train["origin"] == origin
    
    model.fit(X_train, y_train)
    models[origin] = model
```

```python
testing_probs = pd.DataFrame(columns=unique_origins)

for origin in unique_origins:
    
    X_test = test[features]
    
    testing_probs[origin] = models[origin].predict_proba(X_test)[:, 1]
```

```python
testing_probs
```

```python
predicted_origin = testing_probs.idxmax(axis=1)

predicted_origin
```
