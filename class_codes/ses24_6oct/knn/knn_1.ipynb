```python
np.random.seed(2)

dfc = pd.read_csv('dc_airbnb.csv')

temp_x = dfc['price'].str.replace(',', '')
temp_y = temp_x.str.replace('$', '')
dfc['price'] = temp_y.astype('float')

dfc['price'] 

dfc = dfc.loc[np.random.permutation(len(dfc))]

def predict_price(x):
    temp_df = dfc.copy()
    temp_df['distance'] = np.abs(temp_df["accommodates"] - x)
    temp_df = temp_df.sort_values("distance")
    nearest_neighbours = temp_df["price"].head()
    predicted_price = nearest_neighbours.mean()
    return (predicted_price)
```

```python
predict_price(4)
```

```python
predict_price(3)
```

```python
predict_price(10)
```
