```python
path = "https://raw.githubusercontent.com/shobhit-nigam/knowledgeclan/main/datasets/income.csv"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

from sklearn.tree import DecisionTreeClassifier
```

```python
dfa = pd.read_csv(path)
```

```python
str_cols = ['workclass', 'education',
 'marital_status',
 'occupation',
 'relationship',
 'race',
 'sex',
 'native_country',
 'high_income']
```

```python
for name in str_cols:
    col = pd.Categorical(dfa[name])
    dfa[name] = col.codes
```

```python
model = DecisionTreeClassifier(min_samples_split=13, random_state=1)
model.fit(train[list_cols], train["high_income"])

test_predictions = model.predict(test[list_cols])
test_auc = roc_auc_score(test["high_income"], test_predictions)

print(test_auc)
```

```python
train_predictions = model.predict(train[list_cols])
train_auc = roc_auc_score(train["high_income"], train_predictions)

print(train_auc)
```

```python
model = DecisionTreeClassifier(min_samples_split=13, max_depth=7, random_state=1)
model.fit(train[list_cols], train["high_income"])

test_predictions = model.predict(test[list_cols])
test_auc = roc_auc_score(test["high_income"], test_predictions)

print(test_auc)
```

```python
train_predictions = model.predict(train[list_cols])
train_auc = roc_auc_score(train["high_income"], train_predictions)

print(train_auc)
```

```python
model = DecisionTreeClassifier(min_samples_split=80, max_depth=3, random_state=1)
model.fit(train[list_cols], train["high_income"])

test_predictions = model.predict(test[list_cols])
test_auc = roc_auc_score(test["high_income"], test_predictions)

print(test_auc)
```

```python
train_predictions = model.predict(train[list_cols])
train_auc = roc_auc_score(train["high_income"], train_predictions)

print(train_auc)
```
