```python
path = "https://raw.githubusercontent.com/shobhit-nigam/knowledgeclan/main/datasets/Housing.txt"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfa = pd.read_table(path)
```

```python
dfa.head(5)
```

```python
dfa.corr()
```

```python
len(dfa.columns)
```

```python
dummies_dfa = pd.get_dummies(dfa)
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(dummies_dfa.corr(), cmap='YlGnBu', annot=True)
plt.show()
```

```python
import numpy as np

# Create correlation matrix
corr_matrix = dummies_dfa.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features 
dummies_dfa.drop(to_drop, axis=1, inplace=True)
```
