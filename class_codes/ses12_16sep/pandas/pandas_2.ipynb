```python
dictx = {
    'nkey':['N0', 'N1', 'N2','N3'],
    'nkey2':['N0', 'N1', 'N0','N1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
}


dicty = {
    'nkey':['N0', 'N1', 'N2','N3'],
    'nkey2':['N0', 'N0', 'N0','N0'],
    'C': ['A0', 'A1', 'A2', 'A3'],
    'D': ['B0', 'B1', 'B2', 'B3']
}

dfx = pd.DataFrame(dictx)
dfy = pd.DataFrame(dicty)
```

```python
dfx
```

```python
dfy
```

```python
result = pd.merge(dfx, dfy, on=["nkey", "nkey2"])    # how =inner
result
```

```python
result = pd.merge(dfx, dfy, on=["nkey", "nkey2"], how="inner")    # how =inner
result
```

```python
result = pd.merge(dfx, dfy, on=["nkey", "nkey2"], how="left")    # how =left
result
```
