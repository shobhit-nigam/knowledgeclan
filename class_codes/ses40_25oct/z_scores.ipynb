```python
min_val = dfa['SalePrice'].min()
max_val = dfa['SalePrice'].max()
mean_val = dfa['SalePrice'].mean()

def cal_z_score(val, arr, bessel=0):
    mean = sum(arr)/len(arr)
    st_dev = np.std(arr, ddof=bessel)
    distance = val - mean
    z = distance / st_dev
    
    return round(z, 4)
```
